```python
import scipy as sp 
import math as m
import numpy as np 
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns

# Makes the graphs visible in this notebook 
%matplotlib inline 
```

```python
# Apply the default seaborn theme, scaling, and color palette
sns.set()

# Load the test dataset tips 
tips = sns.load_dataset("tips")

tips.info()
tips.describe()
```

```python
tips.head()
```

```python
# Plot the data using relplot (relational plot)

# Display a graph plotting total bill on x axis, tips on y axis, as scatterplot for multiple days
sns.relplot(x = "total_bill", y = "tip", hue = "day", data = tips, kind = "scatter")
```

```python
sns.relplot(x = "total_bill", y = "tip", col = "time", 
          hue = "smoker", style = "smoker", size = "size", 
          data = tips)
```

```python
# Simplest way to create a linear regresion is with regplot (regression plot)
sns.regplot(x = "total_bill", y = "tip", data = tips)
```

```python
# You can also use a lmplot (linear model plot)
sns.lmplot(x = "total_bill", y = "tip", col = "time", hue = "smoker", data = tips);
```

```python
# Other types of plots 

# Histograms 
sns.distplot(tips.total_bill, kde = False, rug = True , bins = 10)
```

```python
sns.distplot(tips.tip, kde = False, rug = True, bins = 6)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
# Generate a random distribution 
x = np.random.normal(size=100)
sns.distplot(x)
```
